This notebook is for getting polygon data of parks from geofabrik dataset.

```python
import geopandas as gpd
import pandas as pd
```

```python
# read in park data
landuse = gpd.read_file('../data/raw-data/washington-latest-free.shp/gis_osm_landuse_a_free_1.shp')
```

```python
landuse 
```

```python
landuse['fclass'].unique()
```

```python
#select park, grass and recreation_ground
parks = landuse[landuse['fclass'].isin(['park', 'grass', 'recreation_ground'])]
parks
```

```python
#save parks to file
parks.to_file('../data/processed-data/parks/parks.shp')
```
