```python
import geopandas as gpd
import pandas as pd
```

```python
#read in clean ped_net data
ped_net = gpd.read_file('../data/cleaned-pednetwork/cleaned-pednetwork-full/cleaned-pednetwork-full.shp')
```

```python
ped_net
```

```python
# select fclass = crossing
crossings = ped_net[ped_net['fclass'] == 'crossing']
```

```python
crossings
```

```python
#read in traffic flow shapefile
traffic_flow = gpd.read_file('../data/Raw_data/2018_Traffic_Flow_Counts-shp/2018_Traffic_Flow_Counts.shp')
```

```python
#convert traffic flow to epsg 2285
traffic_flow = traffic_flow.to_crs(epsg=2285)
```

```python
traffic_flow_base = traffic_flow.explore(tiles='cartodbpositron', column ='AWDT', cmap = 'inferno_r', legend = True, scheme='naturalbreaks')
```

```python
# spatial join crossings and traffic flow
crossings_traffic = gpd.sjoin(crossings, traffic_flow, how='left', predicate='intersects')
```

```python
crossings_traffic
```

```python
#select the rows where pednet_id is duplicated
duplicated = crossings_traffic[crossings_traffic.duplicated(subset='pednet_id', keep=False)]
```

```python
duplicated.explore(tiles='cartodbpositron', m = traffic_flow_base)
```

```python
crossings_traffic.head()
```

```python
#sort by pednet_id and AWDT, include the higher AWDT value
crossings_traffic = crossings_traffic.sort_values(['pednet_id', 'AWDT'], ascending=[True, False])
```

```python
#then drop duplicates, keeping the first value
crossings_traffic = crossings_traffic.drop_duplicates(subset='pednet_id', keep='first')
```

```python
crossings_traffic.head()
```

```python
#keep only the columns we need
crossings_traffic = crossings_traffic[['pednet_id', 'AMPK','PMPK', 'AWDT','ADT','geometry']]
```
