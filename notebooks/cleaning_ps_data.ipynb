This note books is for selecting the Project Sidewalk label types: Curb Ramp, Missing Curb Ramp, Obstacle, Surface Problem. \
As well as cropping the data to the buffered study area.

```python
import geopandas as gpd
import pandas as pd
```

```python
# read in ps data
ps_labels = gpd.read_file('../data/project-sidewalk/attributes_2024-06-20-12_56_15.431/attributes_2024-06-20-12:56:15.431.shp')
```

```python
ps_labels
```

```python
#read in clean ped_net data
ped_net = gpd.read_file('../data/cleaned-pednetwork/cleaned-pednetwork-full/cleaned-pednetwork-full.shp')
```

```python
ped_net
```

```python
# read in study area buffered
study_area_buffered = gpd.read_file('../data/boundaries/study-area-buffered/study-area-buffered.shp')
```

```python
#change ps labels crs to 2285
ps_labels = ps_labels.to_crs(epsg=2285)
```

```python
# filter ps data to only include points within the study area buffer
ps_labels_cropped = gpd.sjoin(ps_labels, study_area_buffered, how='inner', predicate='within')
```

```python
ps_labels_cropped
```

```python
ps_labels_cropped['labelType'].value_counts()
```

```python
#delete labelTypes NoSidewalk, Crosswalk, Occlusion, Signal and Other
ps_labels_cropped = ps_labels_cropped[ps_labels_cropped.labelType != 'NoSidewalk']
ps_labels_cropped = ps_labels_cropped[ps_labels_cropped.labelType != 'Crosswalk']
ps_labels_cropped = ps_labels_cropped[ps_labels_cropped.labelType != 'Occlusion']
ps_labels_cropped = ps_labels_cropped[ps_labels_cropped.labelType != 'Signal']
ps_labels_cropped = ps_labels_cropped[ps_labels_cropped.labelType != 'Other']
```

```python
ps_labels_cropped.head()
```

```python
#save ps labels to a new shapefile
ps_labels_cropped.to_file('../data/project-sidewalk/project-sidewalk-labels.shp')
```
